## 🧠 Part 1: What is an LLM in LangChain?

**LLM = The Brain of your application.**

It reads your text (prompt) and returns a smart response.

```
You give text  →  LLM thinks  →  LLM gives text back
```

### LangChain supports many LLMs:

| LLM | Company | Free? |
|---|---|---|
| GPT-4 | OpenAI | ❌ Paid |
| Gemini | Google | ✅ Free tier |
| Llama 3 | Meta (via Ollama) | ✅ Free |
| Mistral | Mistral (via Ollama) | ✅ Free |

### 🔑 Two types of LLM in LangChain:

| Type | Returns | Usage |
|---|---|---|
| `LLM` (old) | Plain string | Basic use |
| `ChatModel` (modern ✅) | AIMessage object | Use `res.content` to get text |

### 💡 Beauty of LangChain — Swap LLMs easily!
```python
chain = prompt | google_llm    # Google Gemini
chain = prompt | openai_llm   # OpenAI GPT
chain = prompt | ollama_llm   # Local Ollama (FREE!)
```
Just change **one line** — rest of the code stays the same! ✅

````python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_openai import ChatOpenAI

load_dotenv()

# --- Google Gemini LLM ---
google_llm = ChatGoogleGenerativeAI(
    temperature=0,           # 0 = focused answers, 1 = creative answers
    model="gemini-2.0-flash",
    api_key=os.getenv("GOOGLE_API_KEY")
)

# --- OpenAI GPT LLM ---
openai_llm = ChatOpenAI(
    temperature=0,
    model="gpt-4",
    api_key=os.getenv("OPENAI_API_KEY")
)

print("✅ LLMs loaded successfully!")
````

````python
# Direct LLM call (without prompt template)
res = google_llm.invoke("What is AI?")

print(type(res))         # AIMessage object
print(res.content)       # The actual text response
````

## 📝 Part 2: What is PromptTemplate?

**PromptTemplate = A reusable text template with blank spaces (variables).**

Think of it like a letter template:
```
"Dear {name}, You are invited to the party!"
```
You fill in `{name}` later with any value.

### Why use PromptTemplate instead of plain string?

| Plain String | PromptTemplate |
|---|---|
| Fixed text only | Reusable with any input |
| Can't change dynamically | Fill `{variable}` with anything |
| Not flexible | Flexible ✅ |

### `input_variables` — Why is it needed?

It **registers** the blank spaces in your template with LangChain.
```python
input_variables=["input"]   # "Hey LangChain, my template has a variable called 'input'"
```
If you skip it → **Error!** ❌ LangChain won't know what `{input}` is.

````python
from langchain_core.prompts import PromptTemplate

# --- Single variable template ---
prompt = PromptTemplate(
    input_variables=["input"],   # Register the variable
    template="You are a helpful assistant. Complete the user request: {input}"
)

# Preview what the filled prompt looks like
print(prompt.format(input="Where is India located?"))
````

````python
# --- Multiple variables template ---
multi_prompt = PromptTemplate(
    input_variables=["role", "language", "question"],
    template="You are a {role}. Answer this question in {language}: {question}"
)

print(multi_prompt.format(
    role="teacher",
    language="Tamil",
    question="What is AI?"
))
````

---
## 🔗 Part 3: Connecting Prompt + LLM (Chain)

In LangChain, you connect prompt and LLM using the `|` pipe operator:

```
chain = prompt | llm
```

This is called **LCEL (LangChain Expression Language)**.

````python
# Connect prompt + LLM into a chain
chain = prompt | google_llm

# Run the chain
res = chain.invoke({"input": "Where is India located?"})

print(res.content)
````

## 📌 Summary

```
PromptTemplate          LLM                Output
```

| Concept | What it does |
|---|---|
| **LLM** | The brain — reads prompt, gives response |
| **ChatModel** | Modern LLM — returns `AIMessage`, use `.content` |
| **PromptTemplate** | Reusable text template with `{variables}` |
| **input_variables** | Registers blank spaces in the template |
| **chain = prompt \| llm** | Connects everything together (LCEL) |
| **chain.invoke()** | Runs the chain with your input |
